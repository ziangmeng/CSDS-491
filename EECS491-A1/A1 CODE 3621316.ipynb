```python
##Q4.1
import numpy as np
from scipy.stats import beta
import matplotlib.pyplot as plt 

def binomial_posterior(Œò,y, n, a, b):
    post_a = a + y
    post_b = b + n - y
    return beta.pdf(Œò,post_a, post_b)
```

```python
distribution = np.linspace(0,1.0,1000 ) 
y1 = beta.pdf(distribution,7,3) 
plt.plot(distribution, y1)
plt.ylabel('p(Œò)')
plt.xlabel('Œò')
plt.suptitle('I believe strongly that the coin is biased to either mostly heads')
```

```python
from scipy.stats import bernoulli
from scipy.stats import beta
import matplotlib.pyplot as plt 
import numpy as np
from scipy import stats


def plot_binomial_posterior(alpha, beta):
    data = bernoulli(0.5).rvs(100)
    x = np.linspace(0, 1, 100)
    for n in [0, 1, 2, 5, 10, 100]:
        plt.plot(x, stats.beta.pdf(x, alpha + sum(data[:n]), beta + n - sum(data[:n])),label = f'n = {n}')
    plt.legend()
    plt.ylabel('p(Œò)')
    plt.xlabel('Œò')
    plt.show()
```

```python
plt.suptitle('I believe strongly that the coin is biased to either mostly heads')
plot_binomial_posterior(18, 2)
```

```python
plt.suptitle('I believe strongly that the coin is biased to either mostly tails')
plot_binomial_posterior(2, 18)
```

```python
plt.suptitle('I believe strongly that the coin is unbiased')
plot_binomial_posterior(10, 10)
```

```python
plt.suptitle('I dont know anything about the bias of the coin')
plot_binomial_posterior(1, 1)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gamma

# Poisson observations
n = 3

# Observation time
Ts = [0, 0.5, 1.5]

# Prior parameters
alpha = 1
beta = 1

# Posterior parameters
alpha_post = alpha + n
x = np.linspace(0, 10, 1000)
# Plot the posterior distributions
for T in Ts:
    beta_post = beta + T
    plt.plot(x, gamma.pdf(x, alpha_post, scale=1/beta_post), label='T={}'.format(T))

plt.legend()
plt.show()
```
